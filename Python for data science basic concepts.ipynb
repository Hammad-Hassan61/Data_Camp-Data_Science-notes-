# Basic Plots with Matplotlib:
Data visualization is very important to understand the data and its relation.
Matplotlib has a subpackage called pyplot

## Convention
pyplot is import as plt

```python
import matplotlib.pyplot as plt
```

### plt.plot()
It takes two list arguments. The first argument correspond to horizontal and second one to vertical axis. It tells what to plot and how to plot. To make scatter plot use plt.scatter() function.

```python
year = [1950, 1970]
pop = [2.5, 3.6]
plt.plot(year,pop)
plt.scatter(year,pop)
plt.show()
```

### plt.xscale('log')
It will convert x-axis to its logarithim.

```python
year = [1950, 1970]
pop = [2.5, 3.6]
plt.scatter(year,pop)
plt.xscale('log')
plt.show()
```

## Histogram
To create histogram use hist function.
First argument is list of values for which you want to build histogram. 
Second argument is bin, tell the python in how many bins data should be divided. By default it will be 10.

```python
values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.hist(values,bins = 3)
plt.show()
```

# Pandas
Dataframe is the function used to make data in tabular form. We can use dictionaries to form tables. Also, data can be imported from different file formats.

```python
# Pre-defined lists
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]

# Import pandas as pd
import pandas as pd

# Create dictionary my_dict with three key:value pairs: my_dict
my_dict = {'country': names, 'drives_right': dr, 'cars_per_cap': cpc}
# Build a DataFrame cars from my_dict: cars
cars = pd.DataFrame(my_dict)

# Print cars
print(cars)

# Definition of row_labels
row_labels = ['US', 'AUS', 'JPN', 'IN', 'RU', 'MOR', 'EG']

# Specify row labels of cars
cars.index = row_labels

# Print cars again
print(cars)
```

### Importing files to pandas

```python
# Import the cars.csv data: cars
cars = pd.read_csv('cars.csv')

# Print out cars
print(cars)

# To determine the index column
cars = pd.read_csv('cars.csv',index_col = 0)
```

### Selecting columns and Rows
To select only the cars_per_cap column from cars, you can use:

cars['cars_per_cap']
cars[['cars_per_cap']]

The single bracket version gives a Pandas Series, the double bracket version gives a Pandas DataFrame.

```python
# Print out country column as Pandas Series
print(cars['country'])

# Print out country column as Pandas DataFrame
print(cars[['country']])

# Print out DataFrame with country and drives_right columns
print(cars[['country', 'drives_right']])

# Print out first 3 observations
print(cars[:3])

# Print out fourth, fifth and sixth observation
print(cars[3:6])
```

### loc and iloc
With loc and iloc you can do practically any data selection operation on DataFrames you can think of. loc is label-based, which means that you have to specify rows and columns based on their row and column labels. iloc is integer index based, so you have to specify rows and columns by their integer index like you did in the previous exercise.

cars.loc['RU']
cars.iloc[4]

cars.loc[['RU']]
cars.iloc[[4]]

cars.loc[['RU', 'AUS']]
cars.iloc[[4, 1]]

```python
# Print out observation for Japan
print(cars.loc['JPN'])

# Print out observations for Australia and Egypt
print(cars.loc[['AUS' ,'EG' ]])

# Print out drives_right value of Morocco
print(cars.iloc[5,2])
print(cars.loc['MOR', 'drives_right'])

# Print sub-DataFrame
print(cars.iloc[[3,4],[1,2]])
print(cars.loc[['RU','MOR'],['country','drives_right']])
```

It's also possible to select only columns with loc and iloc. In both cases, you simply put a slice going from beginning to end in front of the comma:

cars.loc[:, 'country']
cars.iloc[:, 1]

cars.loc[:, ['country','drives_right']]
cars.iloc[:, [1, 2]]

```python
# Print out drives_right column as Series
print(cars.loc[:, 'drives_right'])
print(cars.iloc[:, 2])

# Print out drives_right column as DataFrame
print(cars.loc[:,['drives_right']])
print(cars.iloc[:,[2]])

# Print out cars_per_cap and drives_right as DataFrame
print(cars.loc[:,['cars_per_cap','drives_right']])
print(cars.iloc[:,[0,2]])
```
